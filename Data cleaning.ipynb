```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\user\Downloads\Customer Call List.xlsx")
df
```

```python
df= df.drop_duplicates()
df
```

```python
# deleting uneccesary symbol from the data

df["Last_Name"]=df["Last_Name"].str.strip("...")
df["Last_Name"]=df["Last_Name"].str.strip("/")
df["Last_Name"]=df["Last_Name"].str.strip("_")
df
```

```python
# removing symbols from the phone number columns

df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3]+ '-' + x[3:6]+ '-' + x[0:3])


df["Phone_Number"]=df["Phone_Number"].apply(lambda x:str(x))
df["Phone_Number"]=df["Phone_Number"].str.replace('nan','')
df["Phone_Number"]=df["Phone_Number"].str.replace('n','')
df["Phone_Number"]=df["Phone_Number"].str.replace('nan','')
df

```

```python
df = df.drop(columns="Not_Useful_Column")
df = df.drop(columns="Phone_Number")
df
```
